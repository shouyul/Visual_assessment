```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('/Users/shouyuling/Library/CloudStorage/OneDrive-UniversityofPittsburgh/Postdoc/Projects/Gensight/Data/VEPTask/Results/AllSub_results2.xlsx')
df.tail(10)
```

```python
data=df[df['FeatureType']=='occi']
data
```

```python
sns.set_theme(style="white",palette='Paired')

g = sns.catplot(data=df[df['FeatureType']=='occi_alpha'],kind='bar', errorbar=None,
            x='Stimuli',y='Accuracy',
            hue='GogglesCondition',col='Subjid')

for ax in g.axes.flatten():
    ax.set_title(ax.get_title(), fontsize=18)  # Adjust subplot titles
    ax.set_xlabel(ax.get_xlabel(), fontsize=14)  # Adjust x-axis labels
    ax.set_ylabel(ax.get_ylabel(), fontsize=14)  # Adjust y-axis labels
    ax.tick_params(labelsize=12)  # Adjust tick labels

plt.show()
```

```python
data=df[(df['Subjid']!=4012) & (df['Stimuli']=='All') & (df['HasOffData']=='Yes')]

sns.set_theme(style="white",palette='Paired')

g = sns.catplot(data=data,kind='bar', errorbar=None,
            y='Accuracy',
            hue='GogglesCondition',hue_order=['GogglesON','GogglesOFF'],
            row='ElecGroup',col='FreqGroup')

for ax in g.axes.flatten():
    ax.set_title(ax.get_title(), fontsize=18)  # Adjust subplot titles
    ax.set_xlabel(ax.get_xlabel(), fontsize=14)  # Adjust x-axis labels
    ax.set_ylabel(ax.get_ylabel(), fontsize=14)  # Adjust y-axis labels
    ax.tick_params(labelsize=12)  # Adjust tick labels
    
g.set(ylim=[15,35])
g.refline(y=25,color='red')

plt.show()
```
